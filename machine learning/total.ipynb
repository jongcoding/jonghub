                - 회귀문제와 분류 문제로 나뉨
                    - 회귀는 목표치가 실수, 분류는 부류값
            - 훈련집합
                - 가로축은 특징, 세로축은 목표치
                - 관측한 4개의 점이 훈련집합을 구서함
                훈련집합 x={x1,x2,..., Xn}, Y={y1,y2,..,Yn}
            - 데이터를 어떻게 모델링할 것인가
                -눈대중으로 보면 직석을 이루므로 직선을 선택하자-> 모델로 직선을 선택한 셈
            - 직선 모델의 수식
                - 2개의 매개변수w와 b
                y=wx+b

            - 기계학습은
                - 가장 정확하게 예측할 수 있는. 즉 최적의 매개변수를 찾는 작업
                - 처음에는 최적값을 모르므로 임의의 값에서 시작하고. 점점 성능을 개선하여 최적에 도달

    특징 공간과 데이터에 대한 이해
        - 1차원과 2차원 특징 공간
            - 1차원 특징공간
            - 2차원 특징공간
                -특징 벡터 표기
                    - x=(x1,x2)T
                - EX
                    - x=(몸무게 키)T, y=장타율
                    - x=(체온, 두통)T, y=감기여부

        - 다차원 특징 공간
            - 다차원 특징공간예제
            - d-차원데이터
                - 특징벡터 표기 x=(x1,x2,...,xd)T
            - d- 차원데이터를 위한 학습 모델
                - 직선 모델을 사용하는 경우 매개변수 수= d+1
                    y=w1x1+w2x2+...+wdxd+b
                - 2차 곡선모델을 사용하면 매개변수 수가 크게 증가
                    - 매개변수 수= d^2+d+1
                    - ex: lris 데이터: d=4이므로 21개의 매개변수
                    - ex: MNIST 데이터: D=784이므로 615,441개의 매개변수
                    y=w1x^2+ w2x2/2+...wdxd^2+ wd+1x1x2+...wd2xd-1xd+wd^2+1x1+...+Wd^2+dxd+b
        - 데이터에 대한 이해
            -> 이해
                - 과학 기술의 발전과정
                    데이터 수집-> 모델 정립-> 예측
                - 기계 학습
                    -기계 학습이 푸는 문제는 훨씬 복잡함
                        ex- '8' 숫자 패턴과 '단추' 패턴의 다양한 변화 양상
                    - 단순한 수학 공식으로 표현 불가능
                    - 자동을 모델을 찾아내는 과정이 필수
            - 데이터 생성과정
                - 데이터 생성 과정을 완전히 아는 인위적 상황의 예제
                    ex- 두 개 주사위를 던져 나온 눈의 합을 x라 할 때, y=(x-7)^2+1점을 받는 게임
                    - 이런 상황을 '데이터 생성 과정을 완전히 알고 있다'고 말함
                        - x를 알면 정확히 y를 예측할 수 있음
                            - 실제 주사위를 던져 x={3,10,8.5}를 얻었다면,y={17,10,2.5}
                        -x의 발생 확률 P(x)를 정확히 알 수 있음
                        - P(x)를 알고 있으므로, 새로운 데이터 생성 가능

                    - 데이터 생성 광정을 알 수 없음

    - 지도 방식에 따른 유형
        - 지도 학습(Supervised Learning)
            - 정답이 주어진 상태에서 학습하는 알고리즘
            - 훈련데이터로부터 하나의 함수를 유추
            - 지도학습 종류
                - 회귀분석(Regression)/예측: 연속적인 값을 출력하는 것
                - 분류(Classifiaciton): 주어진 입력 벡터가 어떤 종류의 값인지 표시하는 것
        - 비지도 학습(Unsupervised Learning)
            - 정답이 주어지지 않은 상태에서 데이터의 특정을 학습하는 알고리즘
            - 데이터가 어떻게 구성되었는지를 알아내는 문제를 해결
            - 지도학습혹은 강화 학습과는 달리 입력값에 대한 목표치(분류된 값)기 없음
            - 비지도학습 종류: 군집화, 연관규칙
                -데이터의 주요 특징을 요약하고 설명
        - 강화학습(Reinforcement Learning)
            - 보상 혹은 벌칙과 함께 여러 번의 시행착오를 거쳐 스스로 학습하는 방법
                - 분류할 수 있는 데이터 혹은 정답이 존재하지 않음
                - 행동에 보상 혹은 벌칙을 받으며 학습
            - 보상을 최대한 많이 얻도록 하는 행동을 유도하도록 학습을 진행

정규분포와 분별함수 베이시언 분류기

```python
import numpy as np

mu = np.array([3, 1])

Sigma = np.array([[2, 0],[0, 2]])

det_Sigma = np.linalg.det(Sigma)
print('Determinant of Sigma=\n', det_Sigma)

inv_Sigma = np.linalg.inv(Sigma)
print('Inverse matrix of Sigma=\n', inv_Sigma)

coef1 = -1/2*inv_Sigma
coef2 = mu.T @ inv_Sigma
coef3 = -1/2*mu.T @ inv_Sigma @ mu - np.log(2*3.14) - 1/2*np.log(det_Sigma)

print('Coefficient 1 =\n', coef1)
print('Coefficient 2 =\n', coef2)
print('Coefficient 3 =\n', coef3)
```

최소위험 베이시언분류기

```python
import numpy as np

# 클래스 파라미터 정의
mu1 = np.array([0, 0, 0])
mu2 = np.array([1, 1, 1])
Sigma1 = np.eye(3)  # 단위 행렬
Sigma2 = np.eye(3)  # 단위 행렬

# 결정 경계 함수
def decision_boundary(mu1, mu2, Sigma1, Sigma2):
    # 공분산 행렬의 역행렬
    inv_Sigma1 = np.linalg.inv(Sigma1)
    inv_Sigma2 = np.linalg.inv(Sigma2)
    
    # 상수항 계산
    constant_term = -0.5 * (np.dot(mu1, np.dot(inv_Sigma1, mu1)) - np.dot(mu2, np.dot(inv_Sigma2, mu2)))
    
    # 결정 경계의 선형 부분 계산
    linear_term = np.dot(inv_Sigma2, mu2) - np.dot(inv_Sigma1, mu1)
    
    return linear_term, constant_term

# 결정 경계 구하기
linear_term, constant_term = decision_boundary(mu1, mu2, Sigma1, Sigma2)
print("결정 경계의 선형 항:", linear_term)
print("결정 경계의 상수항:", constant_term)

# 결정 경계 식: linear_term.T * x + constant_term = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
omega_1 = np.transpose(np.array([[180, 78], [167, 72], [170, 66], [178, 80]])) # 남자
omega_2 = np.transpose(np.array([[160, 48], [155, 45], [167, 55], [158, 60]])) # 여자
X = np.hstack([omega_1, omega_2])
y = np.array([0] * len(omega_1) + [1] * len(omega_2))

# 테스트 샘플
x = np.transpose(np.array([165, 65]))

# 클래스별 평균 벡터 계산
mean_omega_1 = np.mean(omega_1, axis=1)
mean_omega_2 = np.mean(omega_2, axis=1)

# 클래스별 공분산 행렬 계산
cov_omega_1 = np.cov(omega_1)
cov_omega_2 = np.cov(omega_2)

# 역 공분산 행렬 계산
inv_cov_omega_1 = np.linalg.inv(cov_omega_1)
inv_cov_omega_2 = np.linalg.inv(cov_omega_2)

# 평균 벡터 그리기
plt.plot(mean_omega_1[0], mean_omega_1[1], 'ro', markersize=10, label='Man Class Mean')
plt.plot(mean_omega_2[0], mean_omega_2[1], 'bs', markersize=10, label='Woman Class Mean')

# 테스트 샘플 출력
plt.scatter(x[0], x[1], c='g', label='Test Sample')

# 훈련 데이터 출력
plt.scatter(omega_1[0], omega_1[1], c='r', label='Man')
plt.scatter(omega_2[0], omega_2[1], c='b', label='Woman')

x_min, x_max = X[0].min() - 1, X[0].max() + 1
y_min, y_max = X[1].min() - 1, X[1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

Z = np.zeros(xx.shape)

# 두 평균 벡터를 이어주기
plt.plot([mean_omega_1[0], mean_omega_2[0]], [mean_omega_1[1], mean_omega_2[1]], color='black', linestyle='--')

# Mahalanobis 거리 계산 함수
def mahalanobis_distance(x, mean, inv_covariance):
    x_minus_mean = x - mean
    mahala_dist = np.sqrt(np.dot(np.dot(x_minus_mean.T, inv_covariance), x_minus_mean))
    return mahala_dist

# 남자와 여자의 Mahalanobis 거리 계산 및 결정 경계 그리기
for i in range(len(xx)):
    for j in range(len(yy)):
        point = np.array([xx[i, j], yy[i, j]])
        dist_omega_1 = mahalanobis_distance(point, mean_omega_1, inv_cov_omega_1)
        dist_omega_2 = mahalanobis_distance(point, mean_omega_2, inv_cov_omega_2)
        Z[i, j] = dist_omega_1 - dist_omega_2

plt.contour(xx, yy, Z, levels=[0], colors='green')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Decision Boundary with Mahalanobis Distance')
plt.legend()
plt.grid(True)
plt.show()

# 남자 클래스와 여자 클래스의 Mahalanobis 거리 계산
Mahala_dist1 = mahalanobis_distance(x, mean_omega_1, inv_cov_omega_1)
Mahala_dist2 = mahalanobis_distance(x, mean_omega_2, inv_cov_omega_2)

# 남자인지 여자인지 결정
if Mahala_dist1 < Mahala_dist2:
    print("테스트 샘플은 남자에 속합니다.")
else:
    print("테스트 샘플은 여자에 속합니다.")
```

7주차 SVM
- 오류율 최소화
- 여백(margin)을 최대로 하는 평면을 찾음(svm의 차별성)

* 이진 분류를 위한 결정 초평면과 그것의 수학적 특성
d(x)= wTx+b=0
1. d(x)는 전체 특징 공간을 두 영역으로 분할하며 한 쪽 영역세 속하는 점 x는 d(x)>0 이고 다른 쪽에 있는 점은 d(x)<0이다.
2. 하나의 초평면을 표현하는 식은 여럿 있다.(5.1)에 0이 아닌 임의의 상수 c를 곱하여도 같은 초평평면을 나타낸다.
3. w는 초평면의 법선 벡터로서 초평면의 방향을 나타내고 b는 위치를 나타낸다.
4. 임의의 점x에서 초평면까지의 거리
거리: h=|d(x)|/||w||

선형SVM | 선형 분리 가능한 상황
w(직선의 방향)가 주어진 상황에서,
- '두 부류에 대해 직선으로부터 가장 가까운 샘플까지의 거리가 같게 되는'b를 결정
- 여백은 그런 직선에서 가장 가까운 샘플까지 거리의 두 배로 정의함
- 가장 가까운 샘플으 서포트 벡터라 부름

* 문제의 공식화
-> 여백을 가장 크게 하는 결정 초평면 방향, 즉 w를 찾아라
여백= 2h = 2(|d(x)|/||w||)=2/||w||

* 문제를 조건부 최적화 문제로 최적화

wTxi+b>=1, xi w1
wTxi+b<=-1 xi w2
- 훈련 집합 X={(x1,t1),...,(xn,tn)}

* 라그랑제 승수 방법을 기반으로 수식 정리(7주차 p.10)
모든 샘플이 alpha(i)=0 또는 ti(wTxi+b)-1=0이어야 함. alpha(i) not 0인 샘플이 서포트 벡터

```python
from sklearn import svm
import numpy as np

# 데이터 정의
X = np.array([
    [66, 31],  # 샘플 2
    [64, 32],  # 샘플 3 (A)
    [70, 23]   # 샘플 7 (B)
])
y_A = np.array([-1, 1])  # 레이블: 음성(-1), 양성(1) for (A)
y_B = np.array([-1, 1])  # 레이블: 음성(-1), 양성(1) for (B)

# C 값 설정
C = 0.1

# 모델 (A) 학습
model_A = svm.SVC(kernel='linear', C=C)
model_A.fit(X[:2], y_A)

# 모델 (B) 학습
model_B = svm.SVC(kernel='linear', C=C)
model_B.fit(X[[0, 2]], y_B)

# 모델 세부 사항 출력 및 목적 함수 값 계산
def evaluate_model(model, X, y, label):
    # 서포트 벡터의 수
    n_support = len(model.support_)
    
    # 슬랙 변수 계산: max(0, 1 - y_i * (w * x_i + b))
    decision_function = model.decision_function(X)
    margins = y * decision_function  # y_i * (w * x_i + b)
    xi = np.maximum(0, 1 - margins)
    xi_sum = np.sum(xi)  # 슬랙 변수의 합
    
    # 목적 함수 계산
    w_norm = np.linalg.norm(model.coef_)
    objective_value = 0.5 * w_norm**2 + C * xi_sum
    
    # 결과 반환
    return {
        'support_vectors': n_support,
        'w_norm': w_norm,
        'xi_sum': xi_sum,
        'objective_value': objective_value,
        'label': label
    }

result_A = evaluate_model(model_A, X[:2], y_A, "(A)")
result_B = evaluate_model(model_B, X[[0, 2]], y_B, "(B)")

# 결과 출력 및 모델 선택
print(f"Model {result_A['label']} Objective Value: {result_A['objective_value']}")
print(f"Model {result_B['label']} Objective Value: {result_B['objective_value']}")

if result_A['objective_value'] < result_B['objective_value']:
    print(f"Selected Model: {result_A['label']} is better")
else:
    print(f"Selected Model: {result_B['label']} is better")
```

```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

# 주어진 데이터 포인트
X = np.array([[72, 50], [66, 31], [64, 32], [66, 21], [40, 33], [76, 63], [60, 47], [70, 23]])
y = np.array([1, -1, 1, -1, 1, -1, 1, -1])  # 임의의 레이블 설정
sample_numbers = [1, 2, 3, 4, 5, 6, 7, 8]  # 샘플 번호


# 모델 (A) 학습 - 샘플 2와 3
model_A = svm.SVC(kernel='linear', C=0.1)
model_A.fit(X[[1, 2]], y[[1, 2]])

# 모델 (B) 학습 - 샘플 2와 7
model_B = svm.SVC(kernel='linear', C=0.1)
model_B.fit(X[[1, 6]], y[[1, 6]])

# 결정 경계 시각화 함수
def plot_svc_decision_function(model, X, y, ax, title=""):
    """Plot the decision function for a 2D SVC"""
    xlim = (20, 90)
    ylim = (10, 70)
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 50)
    y = np.linspace(ylim[0], ylim[1], 50)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
               s=300, linewidth=1, facecolors='none', edgecolors='red')

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_title(title)

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# 전체 샘플 및 결정 경계 (A)
plot_svc_decision_function(model_A, X, y, axes[0], "Decision Boundary for Model (A)")

# 전체 샘플 및 결정 경계 (B)
plot_svc_decision_function(model_B, X, y, axes[1], "Decision Boundary for Model (B)")

# 샘플 번호 표시
for ax in axes:
    for i, txt in enumerate(sample_numbers):
        ax.annotate(txt, (X[i, 0], X[i, 1]), textcoords="offset points", xytext=(5,-5), ha='center')

plt.show()
```
